```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, MaxAbsScaler, Normalizer
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

### Load the California Housing dataset

```python
california_housing = fetch_california_housing()
X = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)

print("Original Dataset Shape:", X.shape)
print("Original Data Summary:\n", X.describe())
```

### Min-Max Scaling [0,1]

```python
minmax_scaler = MinMaxScaler()
X_minmax = minmax_scaler.fit_transform(X)
```

### Z-Score Standardization (mean=0, std=1)

```python
zscore_scaler = StandardScaler()
X_zscore = zscore_scaler.fit_transform(X)
```

### Robust Scaling (median=0, IQR=1) → Handles Outliers

```python
robust_scaler = RobustScaler()
X_robust = robust_scaler.fit_transform(X)
```

### MaxAbs Scaling (range [-1,1])

```python
maxabs_scaler = MaxAbsScaler()
X_maxabs = maxabs_scaler.fit_transform(X)
```

### L2 Normalization (row-wise normalization)


```python
l2_scaler = Normalizer(norm='l2')
X_l2 = l2_scaler.fit_transform(X)
```

### Log Transformation → Handle Skewed Data

```python
X_log = np.log1p(X)  # log1p handles log(0) safely
```

```python
# Convert all results to DataFrames for comparison
X_minmax_df = pd.DataFrame(X_minmax, columns=X.columns)
X_zscore_df = pd.DataFrame(X_zscore, columns=X.columns)
X_robust_df = pd.DataFrame(X_robust, columns=X.columns)
X_maxabs_df = pd.DataFrame(X_maxabs, columns=X.columns)
X_l2_df = pd.DataFrame(X_l2, columns=X.columns)
X_log_df = pd.DataFrame(X_log, columns=X.columns)

# Display summary of normalized data
print("\nMin-Max Scaling Summary:\n", X_minmax_df.describe())
print("\nZ-Score Scaling Summary:\n", X_zscore_df.describe())
print("\nRobust Scaling Summary:\n", X_robust_df.describe())
print("\nMaxAbs Scaling Summary:\n", X_maxabs_df.describe())
print("\nL2 Normalization Summary:\n", X_l2_df.describe())
print("\nLog Transformation Summary:\n", X_log_df.describe())
```

```python
# Create interactive histogram
feature = 'MedInc'  # Median Income
fig = go.Figure()

fig.add_trace(go.Histogram(x=X[feature], name='Original', opacity=0.5))
fig.add_trace(go.Histogram(x=X_minmax_df[feature], name='Min-Max', opacity=0.5))
fig.add_trace(go.Histogram(x=X_zscore_df[feature], name='Z-Score', opacity=0.5))
fig.add_trace(go.Histogram(x=X_log_df[feature], name='Log', opacity=0.5))

# Layout for better visuals
fig.update_layout(
    title=f"Interactive Comparison of Normalization Methods for {feature}",
    xaxis_title="Value",
    yaxis_title="Count",
    barmode='overlay',  # Overlay histograms
    template='plotly_dark'
)

fig.update_traces(opacity=0.6)  # semi-transparent bars
fig.show()
```
