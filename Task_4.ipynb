```python
complex_json = {
    "company": "TechCorp",
    "founded": 2010,
    "departments": {
        "engineering": {
            "budget": 2000000,
            "employees": [
                {
                    "name": "Alice Chen",
                    "role": "Senior Developer",
                    "projects": [
                        {"name": "Project Alpha", "status": "completed", "budget": 50000},
                        {"name": "Project Beta", "status": "in-progress", "budget": 75000}
                    ]
                },
                {
                    "name": "Bob Kumar",
                    "role": "Team Lead",
                    "projects": [
                        {"name": "Project Gamma", "status": "planning", "budget": 100000}
                    ]
                }
            ]
        },
        "marketing": {
            "budget": 500000,
            "employees": [
                {
                    "name": "Carol Davis",
                    "role": "Marketing Manager",
                    "projects": [
                        {"name": "Campaign X", "status": "completed", "budget": 25000}
                    ]
                }
            ]
        }
    }
}
```

```python
all_employees = []
for dept_name, dept_data in complex_json['departments'].items():
    for employee in dept_data['employees']:
        emp_record = {
            'company': complex_json['company'],
            'company_founded': complex_json['founded'],
            'department': dept_name,
            'department_budget': dept_data['budget'],
            'employee_name': employee['name'],
            'employee_role': employee['role']
        }
        all_employees.append(emp_record)

df4a = pd.DataFrame(all_employees)
```
